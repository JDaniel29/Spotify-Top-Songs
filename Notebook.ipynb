# Top-Songs Analysis

I am going to use Spotipy to do an analysis of my top songs.

```python
# Import Spotipy and Pandas
import spotipy
from spotipy.oauth2 import SpotifyOAuth
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

scope = 'user-top-read '
username = "e76uk01cctjedljb54t61zeqk"
sp = spotipy.Spotify(auth_manager=SpotifyOAuth(scope=scope, username = username))
```

```python
# Get the Top Tracks for all data, the past 6 months, and the past 4 weeks
track_limit = 50

short_term_results = sp.current_user_top_tracks(time_range='short_term', limit=track_limit)
medium_term_results = sp.current_user_top_tracks(time_range='short_term', limit=track_limit)
long_term_results = sp.current_user_top_tracks(time_range='short_term', limit=track_limit)

short_term_dataframe = pd.json_normalize(short_term_results['items'])
medium_term_dataframe = pd.json_normalize(short_term_results['items'])
long_term_dataframe = pd.json_normalize(short_term_results['items'])

display(short_term_dataframe.head())
display(medium_term_dataframe.head())
display(long_term_dataframe.head())
```

```python
# Get the Song Features for each of the songs in each response
short_term_features_response = sp.audio_features(short_term_dataframe["id"])
medium_term_features_response = sp.audio_features(medium_term_dataframe["id"])
long_term_features_response = sp.audio_features(long_term_dataframe["id"])

short_term_features_dataframe = pd.json_normalize(short_term_features_response)
medium_term_features_dataframe = pd.json_normalize(medium_term_features_response)
long_term_features_dataframe = pd.json_normalize(long_term_features_response)

display(short_term_features_dataframe.head())
```

```python
sns.boxplot(
    data=long_term_features_dataframe,
    x = "duration_ms"
)
```
